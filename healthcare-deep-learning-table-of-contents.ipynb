<h2>Executive Summary</h2>


This collection of Kaggle Notebooks is an opinionated introduction to the field of Healthcare Data Science.

Healthcare data scientists are researchers who generate new knowledge from healthcare "Big Data". 

How big is healthcare "Big Data"?  The data warehouse for the United States Department of Veterans Affairs hospital system contains comprehensive clinical information on over 22 million patients.  There are over 2 trillion rows of data stored across 22,000 columns.  

The VA also has large databases of genomic and imaging data.  The publicly-available <a href="https://wiki.cancerimagingarchive.net/display/Public/APOLLO-1-VA" target="_blank">APOLLA-1-VA data set</a>, contains 2.6GB of medical images and 14.7GB of genomic data.  This large amount of data came from just 7 patients!

Many other hospital systems and insurance companies have similarly large data warehouses.  You simply can't analyze that much data on a single computer!

So let's start talking about how to deal with this problem.
